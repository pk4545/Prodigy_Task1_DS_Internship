```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assuming you have already loaded your data
data = pd.read_csv("/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_6298256.csv")

region_counts = data['Region'].value_counts()
bar_width = 1
limit = range(len(region_counts.index))

# Define colors for each bar
colors = ['blue', 'green', 'orange', 'red', 'purple', 'brown']

# Use the 'color' parameter to set the color of each bar
plt.bar(region_counts.index, region_counts.values, color=colors)

plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')

plt.xticks(limit, region_counts.index, rotation=45)
plt.tight_layout()
plt.show()
```
